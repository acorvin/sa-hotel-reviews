```python
# Import libraries

import pandas as pd
import nltk

from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# import nltk
# import ssl

# try:
#     _create_unverified_https_context = ssl._create_unverified_context
# except AttributeError:
#     pass
# else:
#     ssl._create_default_https_context = _create_unverified_https_context

# nltk.download('all')
```

```python
df = pd.read_csv('hotel-reviews.csv', encoding = 'UTF-8')
df
```

```python
# Create the preprocess_text function

def preprocess_text(text):
    
    # Tokenize the text
    tokens = word_tokenize(text.lower())

    # Remove stop words
    filtered_tokens = [token for token in tokens if token not in stopwords.words('english')]

    # Lemmatize the tokens
    lemmatizer = WordNetLemmatizer()

    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]

    # Join the tokens back into a string
    processed_text = ' '.join(lemmatized_tokens)

    return processed_text

df['Description'] = df['Description'].apply(preprocess_text)[:10]
df
```

```python
# Initialize NLTK sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Create the get_sentiment function
def get_sentiment(text):
    if isinstance(text, str):
        scores = analyzer.polarity_scores(text)
        sentiment = 1 if scores['pos'] > 0 else 0
        return sentiment
    else:
        # Handle non-string inputs (e.g., floats) gracefully
        return None

# Apply get_sentiment function
df['sentiment'] = df['Description'].apply(get_sentiment)
df
```

```python
from sklearn.metrics import classification_report

print(classification_report(df['Positive'], df['sentiment']))
```

## Summary
As you can see, the overall accuracy of this rule-based sentiment analysis model is 66%. Further model training and experimentation could achieve a more desired level of performace.
